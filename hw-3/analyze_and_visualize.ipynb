## Сделайте выводы по анализу датасета и построенным графикам. По желанию можете продолжить анализ данных и также отразить это в выводах.

Мы проанализировали датасет Tips, который является частью библиотеки Seaborn. Он включает в себя данные о посещение ресторана, такие как сумма счета, размер чаевых, пол клиента, курящий или некурящий клиент, день недели и прием пищи. Сделаем некоторые выводы на основе проведенного анализа.

#### 1. Средний чек и чаевые.

```python
print('Средний чек =', tips.total_bill.mean().round(2))
print('Средний размер чаевых =', tips.tip.mean().round(2))

tip_percent = (tips.tip / tips.total_bill).mean().round(4) * 100
print(f'В среднем чаевые составляют {tip_percent} % от чека')
```

#### 2. Влияние размера группы на сумму счета и размер чаевых.

```python
bill_tip_bysize = tips.groupby('size')[['total_bill','tip', 'size']].mean()
# процент чаевых от чека
bill_tip_bysize['tip_percent'] = bill_tip_bysize['tip'] / bill_tip_bysize['total_bill'] * 100
# счет на одного человека
bill_tip_bysize['bill_for_one'] = bill_tip_bysize['total_bill'] / bill_tip_bysize['size']

bill_tip_bysize[['total_bill', 'tip', 'bill_for_one', 'tip_percent']]
```

Проанализируем, меняется ли средняя сумма счета или размер чаевых в зависимости от размера группы. 
Логично было бы предположить, что большие группы имеют тенденцию платить больше и оставлять более крупные чаевые.
Однако мы видим, что группы из одного или двух человек имеют более высокий чек и оставляют в качестве чаевых больший процент чека.
Возможно это связано с тем, что чаевые не должны составлять менее определенной суммы, вне зависимости от размера чека.

#### 3. Распределение счетов по дням недели и приему пищи.

```python
sns.catplot(x='day', 
            y='total_bill', 
            data=tips, 
            order=["Thur", "Fri", "Sat", "Sun"], 
            hue='time',
            dodge=True
           );
```

Посмотрим как распределены счета в разные дни недели для разных приемов пищи, из графика видно, что в четверг люди в основном обедают в данном ресторане, а в пятницу, субботу и воскресенье - ужинают. Также мы видим, что в субботу и воскресенье люди чаще ходят в ресторан и тратят больше денег. Из графиков выше мы знаем, что чаевые больше оставляют во время ужина.

#### 4. Влияние пола и привычки курения на размер чаевых.

```python
tips.groupby('sex', observed=False)[['total_bill','tip']].mean()
```

Из таблицы графиков выше мы видим, что в среднем чек мужчин больше чем у женщин, также они оставляют больше чаевых.

```python
tips.groupby('smoker', observed=False)[['total_bill','tip']].mean()
```

Различие в размере чаевых от курящих и не курящих незначительное.
